```python
X = personal_grooming[["A1_01","A1_02", "A1_03", "A1_04","A1_05","A1_06","A1_07","A1_08","A1_09","A1_10","A1_11","A1_12","A1_13","A1_14","A1_15","A1_16","A1_17","A1_18","A1_19","A1_20","A1_21"]]
X
```

```python
#import the sub-module
from sklearn.cluster import KMeans

# instantiate the model with 3 clusters
km = KMeans(n_clusters=5, random_state=1)

#fit the model
km.fit(X)
```

```python
km.labels_
```

```python
# Save the cluster labels
personal_grooming["cluster_std"] = km.labels_
personal_grooming
```

```python
#Counting no. of datapoints in each cluster

personal_grooming["cluster_std"].value_counts()
```
